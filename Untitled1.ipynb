행렬식 계산

```python
import numpy as np

def getMinorMatrix(A, i, j):
    n = len(A)
    M = np.zeros((n-1, n-1))
    for a in range(0, n-1):
        k = a if a < i else a + 1
        for b in range(0, n-1):
            l = b if b < j else b + 1
            M[a, b] = A[k, l]
    return M


def determinant(M):
    M = np.array(M, dtype=float)

    if len(M) == 2:
        return M[0, 0]*M[1, 1] - M[0, 1]*M[1, 0]

    detVal = 0
    for c in range(len(M)):
        detVal += ((-1)**c) * M[0, c] * determinant(getMinorMatrix(M, 0, c))
    return detVal

A = np.array([[-4, 0, 2, -1, 0],
              [ 1, 3,-3, -1, 4],
              [ 2, 0, 1,  3, 0],
              [-2, 1,-3, -1, 5],
              [1, -5, 1, 0, 5]])

print("det(A) =", determinant(A))

A7a = np.array([[ 1,  5, -6],
                [-1, -4,  4],
                [-2, -7,  9]])

A7b = np.array([[ 1,  5, -3],
                [ 3, -3,  3],
                [ 2, 13, -7]])

print("det(7a) =", determinant(A7a))
print("det(7b) =", determinant(A7b))

A8a = np.array([[ 1,  3, 0,  2],
                [-2, -5, 7,  4],
                [ 3,  5, 2,  1],
                [ 1, -1, 2, -3]])

A8b = np.array([[ 1,  3,  3, -4],
                [ 0,  1,  2, -5],
                [ 2,  5,  4, -3],
                [-3, -7, -5,  2]])

print("det(8a) =", determinant(A8a))
print("det(8b) =", determinant(A8b))

A9 = np.array([[ 1,  3, -1,  0, -2],
               [ 0,  2, -4, -1, -6],
               [-2, -6,  2,  2,  9],
               [ 2,  7, -3,  8, -7],
               [ 3,  5,  5,  2,  7]])

print("det(9) =", determinant(A9))
```

역행렬 계산

```python
import numpy as np

def cofactor(A, i, j):
    (n, m) = A.shape
    M = np.zeros((n-1, n-1))
    for a in range(0, n-1):
        k = a if (a < i) else a + 1
        for b in range(0, n-1):
            l = b if (b < j) else b + 1
            M[a, b] = A[k, l]
    return (-1)**(i + j) * np.linalg.det(M)


def inverseByAdjointMatrix(A):
    A = np.array(A, dtype=float)
    detA = np.linalg.det(A)
    (n, m) = A.shape
    adjA = np.zeros((n, m))

    for i in range(0, n):
        for j in range(0, m):
            adjA[i, j] = cofactor(A, i, j)

    if detA != 0.0:
        Ainv = (1.0 / detA) * adjA
        return detA, adjA, Ainv
    else:
        return detA, adjA, None

A = np.array([[-4, 0,  2, -1, 0],
              [ 1, 3, -3, -1, 4],
              [ 2, 0,  1,  3, 0],
              [-2, 1, -3, -1, 5],
              [ 1, -5, 1,  0, 5]])

print("A =\n", A)
detA, adjA, Ainv = inverseByAdjointMatrix(A)
print("\ndet(A) =", detA)
print("\nadj(A) =\n", adjA)
print("\nA inverse =\n", Ainv)

A25a = np.array([[ 2,  1,  3],
                 [-1, -2,  0],
                 [ 2, -2,  1]])

A25b = np.array([[ 2, -3,  1],
                 [ 4,  2,  2],
                 [ 1,  0, -2]])

for name, M in [("25(a)", A25a), ("25(b)", A25b)]:
    detM, adjM, Minv = inverseByAdjointMatrix(M)
    print(f"\n{name} det =", detM)
    print(f"{name} adj(M) =\n", adjM)

A27a = np.array([[ 2, 5, 5],
                 [-1, -1, 0],
                 [ 2, 4, 3]])

A27b = np.array([[1, 0, 1],
                 [0, 1, 1],
                 [2, 0, 1]])

A27c = np.array([[2, 1, 2],
                 [3, 2, 2],
                 [1, 2, 3]])

for name, M in [("27(a)", A27a), ("27(b)", A27b), ("27(c)", A27c)]:
    detM, adjM, Minv = inverseByAdjointMatrix(M)
    print(f"\n{name} det =", detM)
    print(f"{name} adj(M) =\n", adjM)
    print(f"{name} M^{-1} =\n", Minv)
```

벡터연산

```python
import numpy as np

def getVector(mag, deg):
    vec = np.zeros(2)
    vec[0] = mag * np.cos(deg * 2*np.pi / 360)
    vec[1] = mag * np.sin(deg * 2*np.pi / 360)
    return vec

def getMagDeg(vec):
    mag = np.sqrt(vec[0]*vec[0] + vec[1]*vec[1])
    deg = np.arctan(vec[1] / vec[0]) * 360 / (2*np.pi)
    return mag, deg

F1 = getVector(100, 30)
F2 = getVector(120, 60)
Fsum = F1 + F2

magn, angle = getMagDeg(Fsum)
print("결합한 힘의 크기 :", magn)
print("결합한 힘의 방향 :", angle)
```

사잇각과 정사영 계산

```python
import numpy as np

def angle2vectors(v, w):
    vnorm = np.linalg.norm(v)
    wnorm = np.linalg.norm(w)
    vdot  = np.dot(v.T, w)
    cos_th = vdot / (vnorm * wnorm)

    angle = np.arccos(cos_th) * 360 / (2*np.pi)
    return angle, cos_th

def orthProj(u, v):
    uv_dot = np.dot(u, v)
    vv_dot = np.dot(v, v)
    proj = (uv_dot / vv_dot) * v
    return proj

u23a = np.array([2, 4, 1])
v23a = np.array([1, -1, 3])

u23b = np.array([3, 1, -1])
v23b = np.array([2, 1, -3])

for name, u, v in [("23(a)", u23a, v23a),
                   ("23(b)", u23b, v23b)]:
    angle, cos_th = angle2vectors(u, v)
    print(name, "cos θ =", cos_th, ", θ(도) =", angle)

u24a = np.array([2, -1]);       v24a = np.array([1, 3])
u24b = np.array([2, -2, 4]);    v24b = np.array([-1, 1, 2])
u24c = np.array([1, 2, 1, 3]);  v24c = np.array([1, -3, 3, 2])
u24d = np.array([2, -2, 4]);    v24d = np.array([-1, 1, 1])

for name, u, v in [("24(a)", u24a, v24a),
                   ("24(b)", u24b, v24b),
                   ("24(c)", u24c, v24c),
                   ("24(d)", u24d, v24d)]:
    proj = orthProj(u, v)
    print(name, "proj_v u =", proj)
```

```python
import numpy as np

def pprint(msg, A):
    print("---", msg, "---")
    (n, m) = A.shape
    for i in range(0, n):
        line = ""
        for j in range(0, m):
            line += "{0:.2f}".format(A[i, j]) + " "
        print(line)
    print("")

A = np.eye(4)
pprint("A", A)
print("rank(A) =", np.linalg.matrix_rank(A))

B = np.zeros((3, 3))
pprint("B", B)
print("rank(B) =", np.linalg.matrix_rank(B))

C = np.array([[ 2,  5, -3, -4,  8],
              [ 4,  7, -4, -3,  9],
              [ 6,  9, -5,  2,  4],
              [ 0, -9,  6,  5, -6]])
pprint("C", C)
print("rank(C) =", np.linalg.matrix_rank(C))

CT = np.transpose(C)
pprint("C^T", CT)
print("rank(C^T) =", np.linalg.matrix_rank(CT))

A29a = np.array([[1,2,4,4],
                 [3,4,8,0]])
print("29(a) rank =", np.linalg.matrix_rank(A29a))

A29b = np.array([[1,2,3],
                 [2,3,5],
                 [3,4,7],
                 [4,5,9]])
print("29(b) rank =", np.linalg.matrix_rank(A29b))

A30a = np.array([[1,2,1,5],
                 [2,4,-3,0],
                 [-3,1,2,-1],
                 [1,2,-1,1]])
print("30(a) rank =", np.linalg.matrix_rank(A30a))
print("30(a) nullity =", 4 - np.linalg.matrix_rank(A30a))

A30b = np.array([[1,-2,1],
                 [1,-1,3],
                 [1,1,7]])
print("30(b) rank =", np.linalg.matrix_rank(A30b))
print("30(b) nullity =", 3 - np.linalg.matrix_rank(A30b))

A35 = np.array([[2,5,-3,-4,8],
                [4,7,-4,-3,9],
                [6,9,-5,2,4],
                [0,-9,6,5,-6]])
print("35 rank =", np.linalg.matrix_rank(A35))
print("35 nullity =", 5 - np.linalg.matrix_rank(A35))
```
